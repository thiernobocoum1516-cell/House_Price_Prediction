```python
 # ============================================
# MODÈLES AVEC HYPERPARAMÈTRES ÉQUILIBRÉS
# ============================================

import pandas as pd
import numpy as np
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import Lasso, Ridge, ElasticNet
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from xgboost import XGBRegressor
from catboost import CatBoostRegressor
from lightgbm import LGBMRegressor
import warnings
warnings.filterwarnings('ignore')

print("="*60)
print("TEST DE 8 MODÈLES (hyperparamètres équilibrés)")
print("="*60)

# Charger les données
data_path = '../data/'
X_train_selected = np.load(data_path + 'X_train_selected.npy')
X_test_selected = np.load(data_path + 'X_test_selected.npy')
y_train = np.load(data_path + 'y_train.npy')
y_test = np.load(data_path + 'y_test.npy')

print(f"✅ Données chargées !")
print(f"   X_train_selected: {X_train_selected.shape}")
print(f"   X_test_selected: {X_test_selected.shape}")

# ============================================
# MODÈLES AVEC HYPERPARAMÈTRES ÉQUILIBRÉS
# ============================================

models = {
    'Lasso': Lasso(alpha=0.001, max_iter=10000, random_state=42),
    'Ridge': Ridge(alpha=1.0, random_state=42),
    'ElasticNet': ElasticNet(alpha=0.001, l1_ratio=0.5, max_iter=10000, random_state=42),
    'RandomForest': RandomForestRegressor(
        n_estimators=200, max_depth=10, random_state=42
    ),
    'GradientBoosting': GradientBoostingRegressor(
        n_estimators=200, learning_rate=0.05, max_depth=4, random_state=42
    ),
    'XGBoost': XGBRegressor(
        n_estimators=200, learning_rate=0.05, max_depth=4, random_state=42
    ),
    'LightGBM': LGBMRegressor(
        n_estimators=200, learning_rate=0.05, max_depth=4, 
        random_state=42, verbose=-1
    ),
    'CatBoost': CatBoostRegressor(
        iterations=200, learning_rate=0.05, depth=4,
        random_seed=42, verbose=False
    )
}

# Entraînement et évaluation
results = []

for name, model in models.items():
    print(f"🚀 {name}...")
    
    cv_scores = cross_val_score(model, X_train_selected, y_train, cv=5, 
                                scoring='neg_mean_squared_error')
    cv_rmse = np.sqrt(-cv_scores.mean())
    
    model.fit(X_train_selected, y_train)
    y_pred = model.predict(X_test_selected)
    
    val_rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    val_r2 = r2_score(y_test, y_pred)
    
    results.append({
        'Model': name, 
        'CV_RMSE': cv_rmse, 
        'Test_RMSE': val_rmse, 
        'Test_R2': val_r2
    })
    
    print(f"   ✅ CV RMSE: {cv_rmse:.5f} | Test R²: {val_r2:.4f}")

# Classement
print("\n" + "="*60)
print("🏆 CLASSEMENT DES MODÈLES")
print("="*60)

results_df = pd.DataFrame(results).sort_values('Test_R2', ascending=False)

print("-" * 65)
print(f"{'Rang':<5} {'Modèle':<18} {'CV RMSE':<12} {'Test RMSE':<12} {'R²':<8}")
print("-" * 65)

for i, row in results_df.iterrows():
    rank = results_df.index.get_loc(i) + 1
    print(f"{rank:<5} {row['Model']:<18} {row['CV_RMSE']:.5f}    {row['Test_RMSE']:.5f}    {row['Test_R2']:.4f}")

print("-" * 65)

best_model_name = results_df.iloc[0]['Model']
best_r2 = results_df.iloc[0]['Test_R2']

print(f"\n✨ MEILLEUR MODÈLE: {best_model_name} (R² = {best_r2:.4f})")
```

```python
# ============================================
# XGBOOST AVEC RÉGULARISATION RENFORCÉE
# ============================================

from xgboost import XGBRegressor

# Modèle avec régularisation renforcée
best_model = XGBRegressor(
    n_estimators=150,            # ← Réduit
    learning_rate=0.02,          # ← Réduit
    max_depth=3,                 # ← Réduit (profondeur max)
    min_child_weight=5,          # ← Augmenté
    subsample=0.7,               # ← Réduit
    colsample_bytree=0.7,        # ← Réduit
    reg_alpha=0.5,               # ← Ajouté
    reg_lambda=2.0,              # ← Ajouté
    random_state=42
)

# Entraîner
best_model.fit(X_train_selected, y_train)

# Évaluer
y_train_pred = best_model.predict(X_train_selected)
y_test_pred = best_model.predict(X_test_selected)

train_r2 = r2_score(y_train, y_train_pred)
test_r2 = r2_score(y_test, y_test_pred)

train_rmse = np.sqrt(mean_squared_error(y_train, y_train_pred))
test_rmse = np.sqrt(mean_squared_error(y_test, y_test_pred))

print("="*60)
print("XGBOOST AVEC RÉGULARISATION RENFORCÉE")
print("="*60)
print(f"Train R² : {train_r2:.4f}")
print(f"Test R²  : {test_r2:.4f}")
print(f"Écart    : {train_r2 - test_r2:.4f}")
```

```python
# ============================================
# COMPARAISON DE DIFFÉRENTS NOMBRES DE FEATURES
# ============================================
# Au début du notebook de modélisation
X_train_processed = np.load('../data/X_train_processed.npy')
X_test_processed = np.load('../data/X_test_processed.npy')
import numpy as np
import pandas as pd
from sklearn.feature_selection import SelectFromModel
from sklearn.linear_model import Lasso
from xgboost import XGBRegressor
from sklearn.metrics import r2_score, mean_squared_error

print("="*60)
print("COMPARAISON DU NOMBRE DE FEATURES")
print("="*60)

# 1. Tester plusieurs nombres de features
feature_counts = [30, 40, 50, 60, 70, 80]
results = []

for n_features in feature_counts:
    print(f"\n🔍 Test avec {n_features} features...")
    
    # Sélectionner n_features
    lasso = Lasso(alpha=0.001, max_iter=10000, random_state=42)
    selector = SelectFromModel(lasso, max_features=n_features)
    selector.fit(X_train_processed, y_train)
    
    X_train_sel = selector.transform(X_train_processed)
    X_test_sel = selector.transform(X_test_processed)
    
    # Entraîner XGBoost régularisé (le meilleur trouvé)
    model = XGBRegressor(
        n_estimators=150,
        learning_rate=0.02,
        max_depth=3,
        min_child_weight=5,
        subsample=0.7,
        colsample_bytree=0.7,
        reg_alpha=0.5,
        reg_lambda=2.0,
        random_state=42
    )
    
    model.fit(X_train_sel, y_train)
    
    # Évaluer
    y_train_pred = model.predict(X_train_sel)
    y_test_pred = model.predict(X_test_sel)
    
    train_r2 = r2_score(y_train, y_train_pred)
    test_r2 = r2_score(y_test, y_test_pred)
    train_rmse = np.sqrt(mean_squared_error(y_train, y_train_pred))
    test_rmse = np.sqrt(mean_squared_error(y_test, y_test_pred))
    ecart = train_r2 - test_r2
    
    results.append({
        'n_features': n_features,
        'train_r2': train_r2,
        'test_r2': test_r2,
        'ecart': ecart,
        'train_rmse': train_rmse,
        'test_rmse': test_rmse
    })
    
    print(f"   Train R² : {train_r2:.4f}")
    print(f"   Test R²  : {test_r2:.4f}")
    print(f"   Écart    : {ecart:.4f}")
    print(f"   Test RMSE: {test_rmse:.2f}")

# ============================================
# 2. ANALYSE DES RÉSULTATS
# ============================================

print("\n" + "="*60)
print("📊 COMPARAISON DES RÉSULTATS")
print("="*60)

results_df = pd.DataFrame(results)

print("\n📋 Tableau récapitulatif :")
print(results_df.to_string(index=False))

# 3. Trouver le meilleur compromis
print("\n" + "="*60)
print("🎯 MEILLEUR COMPROMIS")
print("="*60)

# Meilleur R² sur test
best_r2 = results_df.loc[results_df['test_r2'].idxmax()]
print(f"\n🏆 Meilleur R² sur test : {best_r2['n_features']} features")
print(f"   Test R² : {best_r2['test_r2']:.4f}")
print(f"   Écart   : {best_r2['ecart']:.4f}")

# Meilleur écart (moins d'overfitting)
best_ecart = results_df.loc[results_df['ecart'].idxmin()]
print(f"\n✅ Meilleur écart Train/Test : {best_ecart['n_features']} features")
print(f"   Test R² : {best_ecart['test_r2']:.4f}")
print(f"   Écart   : {best_ecart['ecart']:.4f}")

# 4. Visualisation
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(results_df['n_features'], results_df['test_r2'], 'bo-', label='Test R²')
plt.plot(results_df['n_features'], results_df['train_r2'], 'ro-', label='Train R²')
plt.xlabel('Nombre de features')
plt.ylabel('R²')
plt.title('R² en fonction du nombre de features')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(results_df['n_features'], results_df['ecart'], 'go-')
plt.xlabel('Nombre de features')
plt.ylabel('Écart Train/Test')
plt.title('Écart R² en fonction du nombre de features')
plt.axhline(y=0.05, color='r', linestyle='--', label='Seuil 5%')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

*"On a  testé différents nombres de features. Bien que 80 features donnent un R² légèrement plus élevé, j## 'ai choisi 60 features car la performance est quasi identique (R² = 0.8475 contre 0.8512) et le modèle est plus simple et plus robuste."*
